## 2. Preparing Data for modelling
### Pre-Processing Steps
#### Encoding
##### Dummy Variable Creation or One Hot Encoding for weathersit variable

```python
# Creating dummy variable for weathersit
weathersit = pd.get_dummies(bike_sharing_df['weathersit'], drop_first=True)
weathersit.head()
```

```python
# Concatinate dummy dataframe to original dataframe
bike_sharing_df = pd.concat([bike_sharing_df, weathersit], axis=1)
```

```python
# Drop base variable of dummy variable
bike_sharing_df.drop('weathersit', axis=1, inplace=True)
```

##### Dummy Variable Creation / One Hot Encoding for season variable

```python
# Create dummy variable for season
season = pd.get_dummies(bike_sharing_df['season'], drop_first=True)
season.head()
```

```python
# Concatinate dummy variable to dataframe
bike_sharing_df = pd.concat([bike_sharing_df, season], axis=1)
```

```python
bike_sharing_df.drop('season', axis=1, inplace=True)
```

##### Trinometry-based Feature Transformation
- Trigonometry-based feature transformation can implemented on any of the cyclical occurring features and avoided one hot encoding for large amount of categorical values
- Apply this to both `mnthday`

Reference: https://towardsdatascience.com/stop-one-hot-encoding-your-time-based-features-24c699face2f

```python
# trigonometry-based feature transformation function
def transformation(column):
  max_value = column.max()
  sin_values = [math.sin((2*math.pi*x)/max_value) for x in list(column)]
  cos_values = [math.cos((2*math.pi*x)/max_value) for x in list(column)]
  return sin_values, cos_values
```

```python
# Feature transformation for Day of Month
bike_sharing_df['mnthday_sin'], bike_sharing_df['mnthday_cos'] = transformation(bike_sharing_df['mnthday'])
```

```python
# Drop base variable for dummy variable
bike_sharing_df.drop('mnthday', axis=1, inplace=True)
```

```python
# Feature transformation for Day of Week
bike_sharing_df['weekday_sin'], bike_sharing_df['weekday_cos'] = transformation(bike_sharing_df['weekday'])
```

```python
# Drop base variable for dummy variable
bike_sharing_df.drop('weekday', axis=1, inplace=True)
```

```python
# Feature transformation for Month
bike_sharing_df['mnth_sin'], bike_sharing_df['mnth_cos'] = transformation(bike_sharing_df['mnth'])
```

```python
# Drop base variable for dummy variable
bike_sharing_df.drop('mnth', axis=1, inplace=True)
```

```python
bike_sharing_df.columns = bike_sharing_df.columns.astype('str')
```

```python
X = bike_sharing_df.drop('cnt', axis=1)
y = bike_sharing_df['cnt']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

```python
# Scaling using Standardization method
X_train = (X_train - X_train.mean()) / X_train.std()
X_test = (X_test - X_test.mean()) / X_test.std()
```
